```python
# We'll be using plotly graph objects :
import plotly.graph_objects as go
fig = go.Figure()
import calendar

df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
filtered_df = df[(df['year'] >= 2015) & (df['year'] <= 2020)]
monthly_data = filtered_df.groupby(['City', 'year', 'month']).agg({
    'CO': 'first',
    'NO2': 'first',
    'NOx': 'first'
}).reset_index()
monthly_data['month'] = monthly_data['month'].apply(lambda x: calendar.month_name[x])
monthly_data = monthly_data[monthly_data['City'] == 'Ahmedabad']

# Will use this color list to give different colors to bars in different frames
color_list = ["#000000", "#1ABC9C", "#2ECC71",  "#3498DB",   "#9B59B6",  "#F1C40F",   "#E67E22",  "#E74C3C",  "#7F8C8D"  ]


def frame_fetch(animated_merged_df, year):

    # Filter the data for year
    animated_merged_df = animated_merged_df[animated_merged_df['year'] == year]

    frame = go.Frame(
        data=[
         go.Bar(x=animated_merged_df['month'], y=animated_merged_df['CO'],marker=dict(color=color_list[np.random.randint(0, 9)]))
        ],
        name=str(year), traces=[0],layout=dict(yaxis=dict(range=[0,animated_merged_df['CO'].max() * 1.2]))
    )

    data = go.Bar(x=animated_merged_df['month'], y=animated_merged_df['CO'],marker=dict(color=color_list[np.random.randint(0, 9)]))

    return frame, data


frames = []
for year in monthly_data['year'].unique():
    frame,data_1 = frame_fetch(monthly_data, year)
    frames.append(frame)

sliders = [
    dict(
        active=0,
    pad = dict(b =10, t =  50),
    len = 0.9,
    x =  0.1,
    y =  0,
    currentvalue =  dict(
        font =  dict(size =  20),
        prefix =  "Year:",
        visible =  True,
        xanchor   = "right"
    ),
        steps=[
            dict(
                method='animate',
                args=[
                    [frame["name"]],
                    dict(

                        frame=dict(duration=500, redraw=False),
                        mode = 'immediate',
                         transition =  dict(duration = 300, easing=  "cubic-in-out")
                    )
                ],
                label=frame["name"]
            ) for frame in frames
        ],


    )
]

layout = go.Layout(
   updatemenus = [
    dict(
        buttons=[
            dict(
                args=[None, {
                    "frame": {"duration": 500, "redraw": False},
                    "fromcurrent": True,
                    "transition": {"duration": 300, "easing": "quadratic-in-out"}
                }],
                label="Play",
                method="animate"
            ),
            dict(
                args=[[None], {
                    "frame": {"duration": 0, "redraw": False},
                    "mode": "immediate",
                    "transition": {"duration": 0}
                }],
                label="Pause",
                method="animate"
            )
        ],
        direction="left",
        pad={"r": 10, "t": 87},
        showactive=False,
        type="buttons",
        x=0.1,
        xanchor="right",
        y=0,
        yanchor="top"
    )
],

    sliders=sliders,  # Make sure to include your sliders definition here
    title="Monthly Carbon Monoxide emission in each year(Ahmedabad)"
)

fig.update(data = [data_1], frames=frames, layout = layout)
fig.update_layout(height = 600, width = 1000)
fig.show()
```
